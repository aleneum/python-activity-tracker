```python
# sorted list of logged activity
seg = data[data.timestamp.between(ts_start, ts_end)]
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(seg[['app', 'title']])
```

```python
# used app sorted by usage duration
seg.groupby("app").duration.sum().sort_values(ascending=False).divide(60).plot(
    label="index", kind="bar", color=plt.cm.gist_rainbow(np.linspace(0, 1.5, seg.app.unique().shape[0])))
plt.ylabel("minutes")
_ = plt.title(f"Used apps from {dt_start.date()} to {dt_stop.date()}")
```
